```python
import matplotlib.pyplot as plt
import numpy as np

size_of_earth=float(input("please enter a value for radius of earth in meters"))
Radius_earth=6400000
Radius_Sun=696000000
Distance=150000000000
Comparable_distance=size_of_earth*Distance/Radius_earth
Comparable_sun_diameter=2*size_of_earth*Radius_Sun/Radius_earth
Comparable_sun_radius=0.5*Comparable_sun_diameter
print("The diameter of sun will be",Comparable_sun_diameter,"meters and will be at a distance of",Comparable_distance,"meters")


fig, ax = plt.subplots()

# Plot the circle at the origin (Earth)
circle1 = plt.Circle((Comparable_sun_radius, 0), 15*Comparable_sun_radius, edgecolor='b', facecolor='green', label='Sun')
ax.add_patch(circle1)

# Plot the circle at the specified distance (Sun)
circle2 = plt.Circle((Comparable_distance, 0), 15*size_of_earth ,edgecolor='r', facecolor='red', label='Earth')
ax.add_patch(circle2)

# Set aspect ratio to be equal
ax.set_aspect('equal', adjustable='box')

# Set labels and title
plt.xlabel('Meters')
plt.ylabel('Meters')
plt.title('Relative Size/Distance between Earth and Sun(radiix15)')

# Set legend
plt.xlim(-size_of_earth, 1.3*Comparable_distance)
plt.ylim(-0.1*Comparable_distance, 0.1*Comparable_distance)
plt.grid(True)


plt.annotate("This is the Sun", # this is the text
                 xy=(Comparable_sun_radius, 0), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(60,0), # distance from text to points (x,y)
                 ha='center',
                 va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.2),
                 arrowprops=dict(arrowstyle = '-|>', connectionstyle='arc3, rad=0'))
plt.annotate("That is the Earth", # this is the text
                 xy=(Comparable_distance, 0), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,15), # distance from text to points (x,y)
                 ha='center',
                 va='bottom',
                 bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.2),
                 arrowprops=dict(arrowstyle = '-|>', connectionstyle='arc3, rad=0'))

plt.show()
#The radii of sun and earth are increased by a factor of 15 to increase visibility.
```
