```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm

```

```python
# Make data
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z=X**(3/2)
```

```python
zlinear = np.arange(0,4,.5)

ylinear = (8 / (8 - 13 * np.sqrt(13))) * (1 - (1 + 9 * ((zlinear)**(2/3)) / 4)**(3/2))

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, vmin=Z.min(), edgecolor='royalblue', lw=0.05, rstride=8, cstride=8,
                alpha=0.1)
ax.scatter(0,ylinear,zlinear, marker='.',c='red')

ax.view_init(elev=30, azim=130, roll=0)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')



plt.show()
```
